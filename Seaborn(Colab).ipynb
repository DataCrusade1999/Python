```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import io
from google.colab import files
sns.set(style='dark')
```

RELPLOT WITH DIFFERENT KINDS

```
df=sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',hue='year',data=df)
```

```
df1=sns.load_dataset('penguins')
sns.relplot(x='body_mass_g',y='species',hue='sex',data=df1)
```

```
df2=sns.load_dataset('mpg')
sns.relplot(x='horsepower',y='acceleration',kind='line',data=df2)
```

```
df3=sns.load_dataset('fmri')
sns.relplot(x='timepoint',y='signal',kind='line',data=df3)
```

```
sns.catplot(x='timepoint',y='signal',kind='box',data=df3)
```

DISTRIBUTION PLOT

```
df4=np.random.randn(1000000)
sns.distplot(df4)
```

FACETGRID

```
var=sns.FacetGrid(df3,col='subject')
var.map(plt.hist,'signal')
```

PAIRGRID

```
var=sns.PairGrid(df3)
var.map(plt.scatter)
```

```
var1=sns.PairGrid(df1)
var1.map_diag(plt.hist)
var1.map_offdiag(plt.scatter)
```

```
var1=sns.PairGrid(df1,hue='sex',palette='Set2',hue_kws={'marker':['o','s']})
var1.map_diag(plt.hist)
var1.map_offdiag(plt.scatter)
var1.add_legend(title='Penguins')
```

```
sns.PairGrid(df1,vars=['culmen_depth_mm','body_mass_g']).map_diag(plt.hist).map_offdiag(plt.scatter)
```

```
sns.PairGrid(df1,x_vars=['body_mass_g','flipper_length_mm'],y_vars=['culmen_depth_mm','culmen_length_mm']).map(plt.scatter)
```
