BASICS

```
import numpy as np
import pandas as pd
import io
import matplotlib.pyplot as plt
from google.colab import files
plt.style.use('seaborn')
```

```
df=np.random.rand(10,10)
df1=np.random.rand(10,10)
plt.scatter(df,df1,edgecolors='black',alpha=0.6,c=np.random.rand(10,10),s=50,cmap='summer')
cbar=plt.colorbar()
cbar.set_label('Colormap')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter')
plt.grid(False)
plt.show()
```

```
df1=[2,3,4,5,6,7]
df2=[5,6,7,8,9,0]
df3=[6,7,7,8,9,7]
#plt.plot(df1,df2,'b--o',label='line1')
plt.plot(df1,df2,color='b',linestyle='--',marker='o',label='line1')
plt.plot(df1,df3,color='r',label='line2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Graph')
plt.legend(loc='upper left')
plt.grid(False)
plt.show()
```

```
data1={'key1':np.arange(0,50),'key2':np.random.rand(50),'key3':np.random.rand(50)}
plt.scatter('key1','key2',data=data1)
plt.scatter('key1','key3',data=data1)
plt.scatter('key2','key3',data=data1)
plt.annotate('fav point',xy=(30,0.8),xytext=(40,0.9),arrowprops=dict(facecolor='green',shrink=0.05))
plt.tight_layout()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(False)
plt.show()
```

```

plt.bar(np.arange(25)+0.25,np.arange(25),width=0.25,color='b',label='bars')
plt.bar(np.arange(25)-0.25,np.arange(25),width=0.25,color='y',label='New Bars')
plt.plot(np.arange(25),np.arange(25),color='k',label='line')
plt.annotate('point',xy=(8,7.5),xytext=(12,14.5),arrowprops={'facecolor':'orange','shrink':0.05})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Chart')
plt.grid(False)
plt.legend()
plt.show()
```

```
df1=pd.read_csv(io.BytesIO(uploaded['houses_to_rent.csv']))
```

```
#Getting Some Important Metrics
print(df1.describe())
```

PLOTTING THE DATA FROM CSV FILE

```
# After performing EDA We Found Correlation Between City And Property Tax
plt.scatter(df1.head(10)['property tax'],df1.head(10)['city'],c=[121,11,89,16,16,28,64,17,26,37],marker='o',alpha=0.6,cmap='summer',edgecolors='black')
cbar=plt.colorbar()
cbar.set_label('Fire Insurance')
plt.xlabel('Property Tax')
plt.ylabel('City Or Not')
plt.title('Houses To Rent')
plt.tight_layout()
plt.grid(False)
plt.show()
```

```
# From EDA We Can See The Correlation Between Rent Amount And Rooms
plt.barh(df1.head(20)['rent amount'],df1.head(20)['rooms'],)
plt.xlabel('Number Of Rooms')
plt.ylabel('Rent Amount')
plt.title('Houses To Rent')
plt.grid(False)
plt.show()
```
