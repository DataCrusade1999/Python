BASICS

```
import numpy as np
import pandas as pd
import io
import matplotlib.pyplot as plt
from google.colab import files
plt.xkcd()

```

```
df=np.random.rand(20,20)
plt.style.use('fast')
plt.scatter(df,np.random.rand(20,20),alpha=0.6,color='k')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter')
plt.grid(False)
plt.show()
```

```
df1=[2,3,4,5,6,7]
df2=[5,6,7,8,9,0]
df3=[6,7,7,8,9,7]
plt.style.use('fast')
#plt.plot(df1,df2,'b--o',label='line1')
plt.plot(df1,df2,color='b',linestyle='--',marker='o',label='line1')
plt.plot(df1,df3,color='r',label='line2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Graph')
plt.legend(loc='upper left')
plt.show()
```

```
data1={'key1':np.arange(0,50),'key2':np.random.rand(50),'key3':np.random.rand(50)}
plt.style.use('tableau-colorblind10')
plt.scatter('key1','key2',data=data1)
plt.scatter('key1','key3',data=data1)
plt.scatter('key2','key3',data=data1)
plt.annotate('fav point',xy=(30,0.8),xytext=(40,0.9),arrowprops=dict(facecolor='green',shrink=0.05))
plt.tight_layout()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(False)
plt.show()
```

BAR CHART

```

plt.style.use('ggplot')
plt.bar(np.arange(25)+0.25,np.arange(25),width=0.25,color='b',label='bars')
plt.bar(np.arange(25)-0.25,np.arange(25),width=0.25,color='y',label='New Bars')
plt.plot(np.arange(25),np.arange(25),color='k',label='line')
plt.annotate('point',xy=(8,7.5),xytext=(12,14.5),arrowprops=dict(facecolor='orange',shrink=0.05))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Chart')
plt.grid(False)
plt.legend()
plt.show()
```

PLOTTING THE DATA FROM CSV FILE

```
# After performing EDA We Found Correlation Between City And Property Tax
plt.style.use('fivethirtyeight')
plt.scatter(df1.head(5)['property tax'],df1.head(5)['city'],marker='o',alpha=0.6)
plt.xlabel('Property Tax')
plt.ylabel('City Or Not')
plt.title('Houses To Rent')
plt.tight_layout()
plt.grid(False)
plt.show()
```

```
# From EDA We Can See The Correlation Between Rent Amount And Rooms
plt.style.use('fivethirtyeight')
plt.barh(df1.head(20)['rent amount'],df1.head(20)['rooms'],)
plt.xlabel('Number Of Rooms')
plt.ylabel('Rent Amount')
plt.title('Houses To Rent')
plt.grid(False)
plt.show()
```
